```python
## Questão 1022

## Sample input: 1 / 2 + 3 / 4
## Sample output: 10/8 = 5/4

## O algoritmo de euclides diz para dividir o maior pelo menor, se o resto for 0,
# então o menor é o mdc, se não então agora o menor será dividido pelo resto e assim sucessivamente até que o resto da divisão seja 0
def algoritmo_euclides(a, b):
    resto = a % b
    valor = b 
    
    if resto == 0: # mdc é o próprio denominador
        if valor < 0:
            return valor*-1
        else:
            return valor
    else:
        while resto != 0:
            novoResto = valor % resto
            
            if novoResto == 0:
                break
            
            valor = resto
            resto = novoResto
        
        if resto < 0:
            return resto*-1
        else:
            return resto

def calcularEquacao():
    equacao = input("Digite a equação: ")

    dados = equacao.split(" ")

    numeros = [int(dados[0]), int(dados[2]), int(dados[4]), int(dados[6])]
    operacao = dados[3]

    if operacao != "/":
        denominador = numeros[1]*numeros[3]
        
        if operacao == "+":
            numerador = numeros[0]*numeros[3] + numeros[1]*numeros[2]
            
        if operacao == "-":
            numerador = numeros[0]*numeros[3] - numeros[1]*numeros[2]

        if operacao == "*":
            numerador = numeros[0] * numeros[2]
    else:
        denominador = numeros[1] * numeros[2]
        numerador = numeros[0] * numeros[3]

    ## Calcular o máximo divisor comum -> Algoritmo de Euclides -> A/B => A = Q*B + R => B/R => ...
    if numerador < denominador:
        mdc = algoritmo_euclides(denominador, numerador)

        print(f"{numerador}/{denominador} = {numerador//mdc}/{denominador//mdc}")
    else:
        mdc = algoritmo_euclides(numerador, denominador)
        print(f"{numerador}/{denominador} = {numerador//mdc}/{denominador//mdc}")
        
chamadas = int(input())

for i in range(0, chamadas):
    calcularEquacao()
```
