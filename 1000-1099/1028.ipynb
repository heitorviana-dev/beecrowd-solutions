```python
## Questão 1028

"""
Input
The first input line contains a single integer N (1 ≤ N ≤ 3000), indicating the number of test cases. Each test case contains two integer numbers F1 (1 ≤ F1 ≤ 1000) and F2 (1 ≤ F2 ≤ 1000)  indicating, respectly, the among of collectable cards that Richard and Vicent have to change.

Output
For each test case there will be an integer number at the output, representing the maximum size of the stack of cards which can be exchanged between two players.
"""

## Pelo o que eu entendi o problema quer que eu retorne a quantidade máxima de cartas cujas pilhas de cartas estarão divididas, por exemplo, 8 e 12 o máximo de cartas de uma pilha seria 4, ou seja, eu quero procurar o mdc entre esses dois valores

def max_cartas_por_pilha(pilha_1, pilha_2): # 100 para ser um valor aleatório
    resto = pilha_1 % pilha_2
    
    if resto == 0:
        return pilha_2
    
    return max_cartas_por_pilha(pilha_2, resto)
    

quantidade_de_casos = int(input("Insira a quantidade de casos de teste: "))

for i in range(quantidade_de_casos):
    pilha_1, pilha_2 = input("Insira a quantidade de cartas de cada pilha: ").split()
    
    if int(pilha_1) > int(pilha_2):
        print(max_cartas_por_pilha(int(pilha_1), int(pilha_2)))
    else:
        print(max_cartas_por_pilha(int(pilha_2), int(pilha_1)))
```
